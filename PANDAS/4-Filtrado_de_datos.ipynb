8 .En este ejercicio tendr√©is que extraer los ataques que han ocurrido por tibur√≥n blanco (lo encontraremos en la columna species como algo que contenga white o White). 


Usad el m√©todo str.contains() para resolver este ejercicio.

```python
patron = "white|White"
```

```python
tiburon_blanco = fichero[fichero["species_"].str.contains(patron, regex = True)]
tiburon_blanco.head(2)
```
