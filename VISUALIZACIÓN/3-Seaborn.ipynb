4 Seleccionad solo los registros de USA, AUSTRALIA y SOUTH AFRICA. Muestra el n칰mero de ataques por pa칤s y a침o desde el a침o 1900.

```python
fichero_uas = fichero[((fichero["country"]== "USA") | (fichero["country"]== "AUSTRALIA") | (fichero["country"]== "SOUTH AFRICA")) & (fichero["year"]>= 1900)]
fichero_uas.sample()
```

```python

plt.figure(figsize= (8,4))

fig = sns.countplot(x= "year", hue="country" ,data= fichero_uas )

fig.set(xlabel= "A침o", ylabel= "N췈 de ataques")

fig.set_title("Ataques pais desde el a침o 1900")

plt.show();
```

```python
# En el gr치fico anterior observamos que hay mayor cantidad en USA en los 칰ltimo sa침os pero decidimos probar con otro gr치fico para entender mejor los datos.

plt.figure(figsize= (20,8))

fig = sns.swarmplot(x= "country", y= "year" ,data= fichero_uas )

fig.set(xlabel= "A침o", ylabel= "N췈 de ataques")

fig.grid()

fig.set_title("Ataques pais desde el a침o 1900")

plt.show();
```

Observamos que los ataques en USA y AUSTRALIA es donde hay m치s ataques a partir del 2000 en USA y en AUTRALIA a partir del 2010, mientras que en SOUTH AFRICA podemos decir que los ataques son bastate homogeneos, entendemos que en este caso es un pa칤s donde el rango de pobreza es mayor por tanto entendemos que exista menos opciones de ocio.
