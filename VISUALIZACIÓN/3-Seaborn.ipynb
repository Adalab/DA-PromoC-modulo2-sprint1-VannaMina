1 Usando la librería seaborn, crea una gráfica de barras que muestre el número de personas que han hecho cada actividad en cada año. Hacedlo solo desde el 2015. Os podrá salir una gráfica un poco fea, no os preocupeis, es lo que esperamos.

```python
# Vamos a filtrar las personas a partir de los años 2015

año_2015 = fichero[fichero["year"]>= 2015]
año_2015.head(2)
```

```python
plt.figure(figsize= (60,15))

fig = sns.countplot(x= "activity", data= año_2015)

fig.set(xlabel= "Actividad", ylabel= "Nº de ataques")

fig.set_title("Actividades por año")

plt.show();
```

Observamos que la mayoría de los casos que hay más ataques es cuando las personas se encuentran haciendo surf, wading, y nadando que son actividades que requieren alejarse más de la costa, por tanto es más facil encontrarse con tiburones.

4 Seleccionad solo los registros de USA, AUSTRALIA y SOUTH AFRICA. Muestra el número de ataques por país y año desde el año 1900.

```python
fichero_uas = fichero[((fichero["country"]== "USA") | (fichero["country"]== "AUSTRALIA") | (fichero["country"]== "SOUTH AFRICA")) & (fichero["year"]>= 1900)]
fichero_uas.sample()
```

```python

plt.figure(figsize= (8,4))

fig = sns.countplot(x= "year", hue="country" ,data= fichero_uas )

fig.set(xlabel= "Año", ylabel= "Nº de ataques")

fig.set_title("Ataques pais desde el año 1900")

plt.show();
```

```python
# En el gráfico anterior observamos que hay mayor cantidad en USA en los último saños pero decidimos probar con otro gráfico para entender mejor los datos.

plt.figure(figsize= (20,8))

fig = sns.swarmplot(x= "country", y= "year" ,data= fichero_uas )

fig.set(xlabel= "Año", ylabel= "Nº de ataques")

fig.grid()

fig.set_title("Ataques pais desde el año 1900")

plt.show();
```

Observamos que los ataques en USA y AUSTRALIA es donde hay más ataques a partir del 2000 en USA y en AUTRALIA a partir del 2010, mientras que en SOUTH AFRICA podemos decir que los ataques son bastate homogeneos, entendemos que en este caso es un país donde el rango de pobreza es mayor por tanto entendemos que exista menos opciones de ocio.

También entendemos que para hacer surf los destinos de USA y AUSTRALIA son más famosos y mucho surfistas acuden a sus playas para realizar la actividad, por tanto también es lógico que gran porcentaje de esas personas puedan ser atacadas. 

También destacamos que por el mismo tema que hemos propuesto en SOUTH AFRICA, en USa hay mayor capacidad de ocio.
 

También observamos que en lo que sería la década de los 70 a los 80 hubo un bajón, que no sabemos muy bien si estaría relacionado con la famosa película "Tiburón".

```python
#Como nos hacía el programa una recomendación para usar el stripplot decidimos probarlo. También hemos probado el catplot pero 
# vemos que los resultados son muy similares así que dejamos el que nos ha recomendado el programa.

plt.figure(figsize= (8,8))

fig = sns.stripplot(x= "country", y= "year" ,data= fichero_uas )

fig.set(xlabel= "Año", ylabel= "Nº de ataques")

fig.set_title("Ataques pais desde el año 1900")

plt.show();
```

5 Realizad un boxplot donde se muestre los principales estadísticos del número de ataques por país.

```python
plt.figure(figsize= (10,20))

fig = sns.boxplot(y= "year", x="country",data= fichero_uas )

fig.set(ylabel= "Año", xlabel="País")

fig.grid())

fig.set_title("Principales estadísticos por país")

plt.show();
```
